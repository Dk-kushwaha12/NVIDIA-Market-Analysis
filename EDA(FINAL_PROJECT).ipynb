```python
## Checking Missing Values
df.isnull().sum()
```

```python
#Filtering DataFrame to show missing values
missing_rows=df[df.isnull().any(axis=1)]
missing_rows
```

```python
#Treating Missing Values with mode using loop
cols_to_impute=["Product Category","Product Name","Region","Customer Segment","Customer Type","Competitor Product"]
for cols in cols_to_impute:
  df[cols].fillna(df[cols].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
# Treating numerical missing values with mean
cols_to_impute = ["Sales Revenue (USD)","Customer Satisfaction","Marketing Spend (USD)","Discount Percentage (%)","Return Rate (%)",
                  "AI/ML Adoption Rate (%)","Ad Campaign Effectiveness","Customer Retention Rate (%)","Market Share (%)"]
for cols in cols_to_impute:
  df[cols].fillna(df[cols].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

##KPI AND VISUALIZATIONS

Q1.Show the average sales revenue (USD) per region.

```python
# Average sales revenue (USD) per region
print("Average sales Revenue (USD) per Region")
print(df.groupby("Region")["Sales Revenue (USD)"].mean().sort_values(ascending=False))
```

Q2.Show the average sales revenue per category

```python
# Average Sale Revenue by Product Category.
print("Average Sales Revenue (USD) Per Product Category")
print(df.groupby("Product Category")["Sales Revenue (USD)"].mean().sort_values(ascending=False))
```

Q3.Which product has the highest sales revenue?

```python
print("Product with Highest Sales Revenue")
print(df.groupby("Product Name")["Sales Revenue (USD)"].sum().sort_values(ascending=False).head(1))
```

Q4.Which product has the lowest sales revenue?

```python
print("Product with Lowest Sales Revenue")
print(df.groupby("Product Name")["Sales Revenue (USD)"].sum().sort_values(ascending=True).head(1))
```

Q5.Show the top 5 Product Name By their Sales Revenue (USD).

```python
# Top 5 Product Name(Total sales revenue)
top_product_category = df.groupby("Product Name")["Sales Revenue (USD)"].sum().sort_values(ascending=False).head(5)
print(top_product_category)
top_product_category.plot(kind="bar",color="teal")
plt.title("Top 5 Product Name by Sales Revenue (USD)")
plt.ylabel("Total Sales")
plt.xlabel("Product Name")
plt.tight_layout()
plt.show()
```

Insights:
 - Top most Product sold is Tesla V100

Q6. Distribution of Market Shares and Product Category.

```python
#Distribution of Market Shares and Product Category
print("Distribution of Market Shares and Product Category")
category_market_share = df.groupby("Product Category")["Market Share (%)"].sum()
print(category_market_share)
```

Q7. How has the "NVIDIA Sales Revenue Trend Over Time"?

```python
df["Product Launch Date"] = pd.to_datetime(df["Product Launch Date"]) # convertion of date into datetime
df["year"]=df["Product Launch Date"].dt.year# Extract year and store into new column
# Line Plot
Sales_per_year=df[["Sales Revenue (USD)","year"]].drop_duplicates()
year_counts = Sales_per_year["year"].value_counts().sort_index()
year_counts.plot(kind="line",marker="o",color="purple")
plt.title("NVIDIA Sales Revenue Trend Over Time")
plt.xlabel("Year")
plt.ylabel("Sales Revenue (USD)")
plt.grid(True)
plt.show()
```

Insights:
1. Overall Stability with Mild Fluctuations
2. sharp rise in revenue between the early 1990s and mid-1990s
3. Peaks in the Late 1990s and Mid-2010s:These might coincide with the launch of major GPU series or growing AI/data center adoption.
4. Recent Decline (Post-2020):

* After 2021, there’s a sharp drop in sales revenue.

* Possible reasons could be:

* Global chip shortages (2021–2022).

Q9.Region-wise sales revenue breakdown.

```python
#pie chart
# Group data by Region and sum the Sales Revenue
region_sales = df.groupby('Region')['Sales Revenue (USD)'].sum()

# Create a pie chart
plt.figure(figsize=(4,4))
plt.pie(
    region_sales,
    labels=region_sales.index,
    autopct='%1.1f%%',
    startangle=140
)
# Equal aspect ratio ensures pie is drawn as a circle
plt.axis('equal')
plt.title('Region-wise Sales Revenue Breakdown')
plt.show()
```

Q12. Discount percentage vs return rate.

```python
df['Discount Bin'] = pd.cut(df['Discount Percentage (%)'], bins=5)
print(df['Discount Bin'])   # showing 5 bins
```

```python
# Box plot

plt.figure(figsize=(6, 3))
sns.boxplot(x='Discount Bin', y='Return Rate (%)', data=df)
plt.title('Box Plot: Discount Bins vs Return Rate')
plt.xlabel('Discount Percentage Bins')
plt.ylabel('Return Rate (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Insights:
- There no outliers shown in the dataset.
